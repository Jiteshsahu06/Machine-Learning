# Step-2 Random Over Sampling

```python
x =dff[["Age","EstimatedSalary"]]
y=dff["Purchased"]
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ro =RandomOverSampler()
ro_x,ro_y=ro.fit_resample(x,y)
```

```python
ro_x.value_counts()
```

```python
ro_y.value_counts()
```

```python
x_train,x_test, y_train  ,y_test = train_test_split(ro_x,ro_y,random_state=42,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg1 =LogisticRegression()
lg1.fit(x_train,y_train)
```

```python
lg1.score(x_test,y_test)*100
```

```python
lg1.predict([[19,19000]])   # answers is 0
```

```python
lg1.predict([[50,20000]])  # answer is 1
```

# Step-3 Random under Sampling

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
ru = RandomUnderSampler()
ru_x,ru_y =ru.fit_resample(x,y)
```

```python
ru_x.value_counts()
```

```python
ru_y.value_counts()
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test, y_train  ,y_test = train_test_split(ru_x,ru_y,random_state=42,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg2 =LogisticRegression()
lg2.fit(x_train,y_train)
```

```python
lg2.score(x_test,y_test)*100
```

```python
lg2.predict([[19,19000]])
```

```python
lg2.predict([[50,20000]])
```
