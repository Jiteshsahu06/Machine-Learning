# Step-1 Logistic Regression (Binary Classification) (Single input or Single output)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df =pd.read_csv(r"C:\Users\jites\OneDrive\Desktop\archive\Social_Network_Ads.csv")
df.drop(columns="EstimatedSalary",inplace=True)      # drop column because we work on logistic regression (Binary classification)
print(df.head(3))
print()
print(df.shape)
```

## A. Cheak dataset is Linearly Separable OR not 

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="Age",y="Purchased",data=df)
plt.show()
```

```python
x= df[["Age"]]                              # Single Input
y=df["Purchased"]                           # Single output
```

## B. Split Dataset in Train test 

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
x_train.shape,x_test.shape
```

## C. Apply Logistic Regression Algorithm

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr =LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[19]])
```

## D. Create s-curve in graph

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="Age",y="Purchased",data=df)
sns.lineplot(x="Age",y=lr.predict(x),data=df,c="red")
plt.show()
```

# Step-2 Logistic Regression (Binary Classification) (Multiple input)

```python
dataset =pd.read_csv(r"C:\Users\jites\OneDrive\Desktop\archive\Social_Network_Ads.csv")
dataset.head(3)
```

## A. Cheak dataset is Linearly Separable OR not 

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="Age",y="EstimatedSalary",data=dataset ,hue="Purchased")
plt.show()
```

```python
x1=dataset.iloc[:,:-1]
y1= dataset["Purchased"]
```

## B. Split Dataset in Train test 

```python
from sklearn.model_selection import train_test_split
```

```python
x_train ,x_test,y_train,y_test =train_test_split(x1,y1,test_size=.2,random_state=42)
```

```python
x_train.shape ,x_test.shape
```

## C. Apply Logistic Regression Algorithm

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr1= LogisticRegression()
lr1.fit(x_train ,y_train)
lr1.score(x_test,y_test)*100
```

```python
lr1.predict([[47, 20000]])
```

## D. Create s-curve in graph by mlxtend library

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x1.to_numpy(),y1.to_numpy(),clf=lr1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr5 =LogisticRegression()
lr5.fit(x_train,y_train)
```

```python
lr5.score(x_test,y_test)*100
```

```python
lr5 =LogisticRegression(multi_class='ovr')
lr5.fit(x_train,y_train)
```

```python
lr5.score(x_test,y_test)*100
```

## Method 2:- Multinomial method

```python
lr5 =LogisticRegression(multi_class='multinomial')
lr5.fit(x_train,y_train)
```

```python
lr5.score(x_test,y_test)*100
```
