# Feature Selection

#### The goal of feature selection techniques in machine learning is to find the best set of features that allows one to build optimized models of studied phenomena.
#### ** Types of Feature Selection Techniques in ML

#### 1. Forward Elimination
#### 2. Backward Elimination

# Step-1. Forward Elimination
# Step-2. Backward Elimination
# 

```python
import pandas as pd
from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
dataset =pd.read_csv(r"C:\Users\jites\OneDrive\Desktop\archive\diabetes.csv")
dataset.head(3)
```

# Step-1. Forward Elimination

```python
dataset.shape
```

```python
x =dataset.iloc[:,:-1]
y =dataset["Outcome"]
```

```python
from sklearn.linear_model import LogisticRegression 
```

```python
logi = LogisticRegression()
```

```python
featureselection =SequentialFeatureSelector(logi ,k_features=4, forward=True)
featureselection.fit(x,y)
```

```python
featureselection.feature_names
```

```python
featureselection.k_feature_names_
```

```python
featureselection.k_score_
```

# Step-2. Backward Elimination

```python
dataset.shape
```

```python
x =dataset.iloc[:,:-1]
y =dataset["Outcome"]
```

```python
from sklearn.linear_model import LogisticRegression 
```

```python
logi = LogisticRegression()
```

```python
featureselection =SequentialFeatureSelector(logi ,k_features=4, forward=False)
featureselection.fit(x,y)
```

```python
featureselection.feature_names
```

```python
featureselection.k_score_
```
